```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tracks = pd.read_csv('metadata/tracks.csv',low_memory = False)
tracks.columns
```

```python
tracks.columns = [tracks.iloc[0,:]]
tracks = tracks.drop([0,1])
```

```python
x = list(tracks.columns[1:])
x.insert(0,'track_id')
x
tracks.columns = x
```

```python
tracks.track_id = pd.to_numeric(tracks.track_id)
```

```python
genres = pd.read_csv('metadata/genres.csv',low_memory = False)
genres.genre_id = pd.to_numeric(genres.genre_id)
genres
```

```python
features = pd.read_csv('metadata/features.csv',low_memory = False)
features.head()
```

```python
features = features.drop([0,1,2])
x = list(features.columns[1:])
x.insert(0,'track_id')
features.columns = x
```

```python
features.track_id = pd.to_numeric(features.track_id)
```

```python
features.head()
```

```python
echonest = pd.read_csv('metadata/echonest.csv',low_memory = False)
echonest.head()
```

```python
echonest.columns = echonest.iloc[1,:]
echonest = echonest.drop([0,1,2])
x = list(echonest.columns[1:])
x.insert(0,'track_id')
echonest.columns = x
echonest.head()
```

```python
echonest.track_id = pd.to_numeric(echonest.track_id)
```

```python
with open("SongList.txt") as file:   
    song_list = file.read()  
type(song_list)
song_list= list(song_list.split('\n'))
new_song_list = []
for i in song_list:
    i = int(str(i[:-4]))
    new_song_list.append(i)
song_list = new_song_list
song_list
```

```python
new_tracks = tracks[tracks["track_id"].isin(song_list)]
len(new_tracks)
new_features = features[features["track_id"].isin(song_list)]
len(new_features)
new_echonest = echonest[echonest["track_id"].isin(song_list)]
len(new_echonest)
new_tracks.to_csv('new_tracks.csv')
new_features.to_csv('new_features.csv')
new_echonest.to_csv('new_echonest.csv')
```

```python
genre = set(new_tracks["genre_top"])
genre = list(genre)
```

```python
new_genres = genres[genres['title'].isin(genre)]
new_genres.to_csv('new_genres.csv')
```
